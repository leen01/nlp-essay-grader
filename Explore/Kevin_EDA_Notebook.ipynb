```
import pandas as pd
import numpy as np
import seaborn as sns
```

```
df = pd.read_csv('data/train.csv')

print(f'Columns: {df.columns}')

df.hist(column='cohesion')
df.hist(column='syntax')
df.hist(column='vocabulary')
df.hist(column='phraseology')
df.hist(column='grammar')
df.hist(column='conventions')
```

```
print(df['vocabulary'].value_counts())
```

```
scored5 = df[df['vocabulary'] == 5]

scored5
```

```
df['total'] = df['cohesion'] + df['syntax'] + df['vocabulary'] + df['phraseology'] + df['grammar'] + df['conventions']
df['average'] = df['total'] / 6
df.head()
df['average'].value_counts()
df.hist(column='average')
```

```
df['highest'] = df[['cohesion', 'syntax', 'vocabulary', 'phraseology', 'grammar', 'conventions']].max(axis=1)
df['truncate'] = df['highest'].round(0)

df.head()
```

```
sns.histplot(
    df,
    x="average", hue="truncate",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
).set(xlabel='Average Score')
```

```
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```
# Compare token usage across examples
# What does the number of commas, apostrophes, semi colons, ect. say about a text's grammar?
```

```
def token_counter(tokens: list, text: str) -> dict:
    counts = {token:0 for token in tokens}

    for character in text:
        if character in counts:
            counts[character] += 1
    
    return counts

token_counter(['!'], "this is a test!")

df['tokens'] = df['full_text'].apply(lambda x: token_counter(['!', '.', ',', ';', ':', '`', '-'], x))
df.head()
```

```
tokens_df = df['tokens'].apply(pd.Series)
tokens_df
```

```
tokens_df['average'] = df['average']
tokens_df['highest'] = df['highest']
tokens_df['truncate'] = df['truncate']
tokens_df['syntax'] = df['syntax']
tokens_df
```

```
sns.histplot(
    tokens_df,
    x=".", hue="truncate",
    multiple="stack",
    edgecolor=".3",
    palette="light:m_r",
    linewidth=.5,
)
```

```
sns.histplot(
    tokens_df,
    x=",", hue="syntax",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
)
```
